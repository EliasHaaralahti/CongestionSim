```python
import matplotlib.pyplot as plt
import numpy as np
import random

x = np.linspace(0,2*np.pi,10)
y = np.sin(x)
```

```python
from matplotlib.path import Path
from markers import get_car_marker, get_human_marker

human_marker = get_human_marker()
car_marker = get_car_marker()


for i in range(len(x)):
    chance = random.uniform(0, 1)
    if chance > 0.5:
        plt.plot(x[i], y[i], marker=human_marker, markersize=20, c='g')
    else: 
        if chance < 0.25:
            plt.plot(x[i], y[i], marker=car_marker, markersize=20, c='b')
        else:
            plt.plot(x[i], y[i], marker=car_marker, markersize=20, c='r')

# plot the actual coordinate for measuring centerness
plt.plot(x, y, 'o', markersize=1, c='k')
```
