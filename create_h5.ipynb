```python
import numpy as np
import h5py
from matplotlib import pyplot as plt

# Generate a batch of 10 RGB images of size 520x520
random_images = np.random.randint(low=0, high=255,size=(10,520,520,3), dtype=np.uint8)
plt.imshow(random_images[0])
plt.show()
```

```python
with h5py.File('test.h5', 'w') as hdf:
    hdf.create_dataset('images', data=random_images)
```
